<h1>Lesson 5: Sets and tuples</h1>

<h3>Sets</h3>

<p>The sets are collections of items defined by the <code>{}</code> brackets
<p>Properties of the sets:</p>
<ul><li>unordered</li><li>unindexed</li><li>Once a set is created, you cannot change its items, but you can remove items and add new items.</li><li>do not allow duplicates</li></ul>
<p>The example of the set is below:</p>

```python
this_set = {"Lama", "Alpaca", "Camel", "Shark"}
```

<p>You can not acess the items in the set by index, but you can loop through them using <code>for</code> loop. This is very similar to looping through the list.</p>

```python
my_set = {"Lama", "Alpaca", "Camel", "Shark"}
for item in my_set:
    print(item)
```

<p>Trying to acess the item by index will not work, because items within the sets <b>do not have indexes.</b></p>

```python
my_set[2]
```

<h3>Measuring length of the set using <code>len()</code> method</h3>


```python
my_set = {"Lama", "Alpaca", "Camel", "Shark"}
print(len(my_set))
```

<h3>Creating the sets from the list or any other iterable</h3>

<p>Below is the example of set creation from the list.</p>

```python
my_list = ["Lama", "Alpaca","Shark", "Camel", "Shark"]
my_awsome_set = set(my_list)
print(my_awsome_set)
```

<p>If you want to create an empty set, also use a <code>set()</code> synthax. If you just make a curly brackets <code>{}</code>, it is a dictionary. </p>

```python
empty_set = set()
print(type(empty_set))
empty_dictionary = {}
print(type(empty_dictionary))
```

<h3>Things you can store in sets</h3>

<p>The sets only can store immutable elements. That is : </p>
<ul><li>Strings</li><li>floats</li><li>tuples if these tuples contain immutable elements</li>
<li>other sets</li></ul>
<p>But not lists and dictionaries</p>

```python
empty_set = set() # creating the empty sets 
empty_set.add("this is string")
print(empty_set)
empty_set.add(["this", "is", "list"])
```

```python
empty_set = set()
empty_set.add({"dict_key":"dict:val"})
```

### Things you can do with sets

Adding item to the set

```python
my_set = {"Lama", "Alpaca", "Camel", "Shark"}
my_set.add("Guanako")
print(my_set)
```

<h4>Merging two sets</h4>

Merging two sets using <code>.update()</code> method, also can be written as <code>|= </code>. In this example yoiu are adding the contents of the <code>my_set2</code> to <code>my_set1</code>.

```python
my_set1 = {"Lama", "Alpaca"}
my_set2 = {"Camel", "Shark"}
my_set1.update(my_set2)
print(my_set1)
```

```python
my_set1 = {"Lama", "Alpaca"}
my_set2 = {"Camel", "Shark"}
my_set1 |= (my_set2)
print(my_set1)
```

<p>You can use this method to add just one item</p>

```python
my_set1 = {"Lama", "Alpaca"}
my_set1.update({"Guanako"})
print(my_set1)
```

<p>Sets <b>do not allow</b> the dupliactes. So if you are trying to merge two sets that share items, only one instance will be added</p> <p>Example below: </p>

```python
my_set1 = {"Lama", "Alpaca", "Camel"}
my_set2 = {"Shark", "Alpaca"}
my_set1.update(my_set2)
print(my_set1)
```

<p> You can also add list or any other itearble (e.g tuple) in the set. </p>

```python
my_set1 = {"Lama", "Alpaca", "Camel"}
list_to_add = ["Guanako", "shark"]
my_set1.update(list_to_add)
print(my_set1)
```

<p>Joining two sets using <code>union()</code> method. Unlike <code>update()</code> method it does not update one of the sets by adding another, but creates the third set containing a merger of two sets</p>

```python
my_set1 = {"Lama", "Alpaca", "Camel"}
my_set2 = {"Shark", "Alpaca"}
joined_set = my_set1.union(my_set2)
print(joined_set)
```

<p>Duplicates in the sets are not allowed, so if the element repeats, it gets added to the joined set only once.</p>

```python
my_set1 = {"Lama", "Alpaca", "Camel"}
my_set2 = {"Shark", "Alpaca", "Camel"}
joined_set = my_set1.union(my_set2)
print(joined_set)
```

<p>You can add join multiple sets using <code>union()</code>.</p>

```python
my_set1 = {"Lama", "Alpaca", "Camel"}
my_set2 = {"Shark", "Alpaca"}
my_set3 = {"Guanako", "Zebrafish"}
long_joined_set = my_set1.union(my_set2, my_set3)
print(long_joined_set)
```

<p>You can use <code>|</code> instead of the <code>union</code></p>

```python
my_set1 = {"Lama", "Alpaca", "Camel"}
my_set2 = {"Shark", "Alpaca", "Camel"}
joined_set = my_set1 | my_set2
print(joined_set)
```

<p>You can use <code>|</code> to join multiple sets </p>

```python
my_set1 = {"Lama", "Alpaca", "Camel"}
my_set2 = {"Shark", "Alpaca"}
my_set3 = {"Guanako", "Zebrafish"}
joined_long_set = my_set1 | my_set2 | my_set3
print(joined_long_set)
```

<p>Removing the item from set using either <code>remove</code> or <code>discard</code> methods</p>

```python
my_set = {"Lama", "Alpaca", "Camel", "Shark"}
my_set.remove("Shark")
print(my_set)
```

<p>If the element you are trying to remove is not in the set, the <code>remove</code> method will raise an error</p>

```python
my_set = {"Lama", "Alpaca", "Camel", "Shark"}
my_set.remove("Zebrafish")
```

```python
my_set = {"Lama", "Alpaca", "Camel", "Shark"}
my_set.discard("Zebrafish")
print(my_set)
```

```python
my_set = {"Lama", "Alpaca", "Camel", "Shark"}
my_set.discard("Alpaca")
print(my_set)
```

<h3>Other operations with sets</h3>

<p><code>intersection()</code> method returns intersection between two sets, that is elements that are shared between the compared sets.</p>

```python
my_set1 = {"Lama", "Alpaca", "Camel"}
my_set2 = {"Shark", "Alpaca"}
intersection_set = my_set1.intersection(my_set2)
print(intersection_set)
```

<p>You can use <code>&</code> instead of the <code>intersection()</code></p>

```python
my_set1 = {"Lama", "Alpaca", "Camel"}
my_set2 = {"Shark", "Alpaca", "Camel"}
```

```python
intersection_set1 = my_set1 & my_set2
print(intersection_set1)
```

<p>You can find intersection of more than two sets as well</p>

```python
my_set1 = {"Lama", "Alpaca", "Camel"}
my_set2 = {"Shark", "Alpaca", "Camel"}
my_set3 = {"Lama", "Alpaca", "Zebra"}
intersection_of_3 = my_set1 & my_set2 & my_set3
print(intersection_of_3)
```

<p>The code below is the same as above, just different syntax</p>

```python
my_set1 = {"Lama", "Alpaca", "Camel"}
my_set2 = {"Shark", "Alpaca", "Camel"}
my_set3 = {"Lama", "Alpaca", "Zebra"}
intersection_of_3 = my_set1.intersection(my_set2, my_set3)
print(intersection_of_3)
```

<code>difference()</code> or <code>-</code>. Results in a set that is a difference of two or more sets. That is, return the items that are <b>present only in the first set</b>, but are not in second. 

```python
my_set1 = {"Lama", "Alpaca", "Camel"}
my_set2 = {"Shark", "Alpaca", "Camel"}
diff_set = my_set1.difference(my_set2)
print(diff_set)
```

```python
my_set1 = {"Lama", "Alpaca", "Camel"}
my_set2 = {"Shark", "Alpaca", "Camel"}
diff_set1 = my_set1 - my_set2
print(diff_set1)
```

```python
my_set1 = {"Lama", "Alpaca", "Camel", "Guanako"}
my_set2 = {"Shark", "Alpaca", "Camel"}
my_set3 = {"Lama", "Alpaca", "Zebra"}
resultant_set = my_set1 - my_set2 - my_set3
print(resultant_set)
```

<p><code>symmetric_difference()</code> or <code>^</code>. This operation creates the third set that contains all the elements that do not repeat from both sets.</p>

```python
my_set1 = {"Lama", "Alpaca", "Camel", "Guanako"}
my_set2 = {"Shark", "Alpaca", "Camel"}
symmetric_set = my_set1.symmetric_difference(my_set2)
symmetric_set
```

```python
my_set1 = {"Lama", "Alpaca", "Camel", "Guanako"}
my_set2 = {"Shark", "Alpaca", "Camel"}
symmetric_set1 = my_set1 ^ my_set2
print(symmetric_set1)
```

<p><code>symmetric_difference_update</code> or <code>^=</code> is very similar to the <code>symmetric_difference</code>, except it does not create the third set, it updates the first of the two sets involved in the operation.</p>

```python
set1 = {"Lama", "Alpaca", "Camel", "Guanako"}
set2 = {"Shark", "Alpaca", "Camel"}
set1.symmetric_difference_update(set2)
print(set1)
```

<p><code>clear()</code> method can be used to empty the set. When you print empty set you see <code>set()</code> instead of the <code>{}</code>, because <code>{}</code> is reverved for empty dictionary</p>

```python
set2 = {"Shark", "Alpaca", "Camel"}
set2.clear()
print(set2)
set2.add("Guanako")
print(set2)
```

<h3>Tuples</h3>

<ul><li>Tuples are one of 4 collection types available in python. (Others are dictionaries, lists and sets)</li>
    <li>Tuples are defined by the <code>()</code> brackets</li>
    <li>Tuples are unchangeable, meaning, once the tuple is created, it can not be changed.</li>
    <li>Elements within tuples have indexes.</li>
    <li>Tuples alllow duplicate values</li>
</ul>

<p>Acessing elements within the tuple and slicing the tuple is similar to that of list</p>

```python
my_tuple = ("Llama", "Camel", "Alpaca")
print(my_tuple[1])
```

<p>Slicing the tuple is similar to that of the list. The end coordinate is not indvcluded. The numbering starts with 0.</p>

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark")
print(my_tuple[1:3])
```

<p>Unlike with the list you can not just re-assign the value of the element, because tuples are unchangeable. The line below will not work.</p>

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark")
my_tuple[1] = "Zebra"
```

<p>But there is a work-around <b>You can convert the tuple to list</b> change it and then convert back</p>

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark")
my_list = list(my_tuple)
print(my_list)
my_list[1] = "Guanako"
convertedBack = tuple(my_list)
print(convertedBack)
```

<p>You can update the tuple using concatenation. This is similar to updating the string, technically what happens a new tuple is created and stored in the variable. </p>
<p>By the way if you are defining the tuple with one element e.g <code>("Shark", )</code> it still needs to have <code>,</code> in it.</p>

```python
my_awsome_tuple = ("Llama", "Camel", "Alpaca")
my_awsome_tuple += ("Shark", )
print(my_awsome_tuple)            
```

<p>Concatenating tuples</p>

```python
my_tuple = ("Llama", "Camel")
my_second_tuple = ("Alpaca", "Shark")
my_entire_tuple = my_tuple + my_second_tuple
print(my_entire_tuple)
```

<p><b>Unpacking tuples </b></p>
<p>The easiest ways to unpack the tuple is to create as many variables as there are elements in the tuple. See example below</p>

```python
my_awsome_tuple = ("Llama", "Camel", "Alpaca")
(first_elem, second_elem, third_elem) = my_awsome_tuple
print(first_elem)
```

<p>Unpacking tuple with asterisk <code>*</code>. Asterisk puts all the values that weren't allocated to the variables in a list.</p>

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark")
(first_elem, *other_elems, last_elem) = my_tuple
print(first_elem)
print(other_elems)
print(last_elem)
```

<p>In the example below, the first and the two last elements are stored in the variables and the rest is stored in the variable with <code>*</code> star</p>

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark", "Guanaco", "Vicuna")
(first_elem, *other_elems, before_last_elem, last_elem) = my_tuple
print(first_elem)
print(other_elems)
print(before_last_elem)
print(last_elem)
```

<p>You can also loop through tuple.</p>

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark", "Guanaco", "Vicuna")
for elem in my_tuple:
    print(elem)
```

<h3>Tuple methods</h3>

<p><code>count()</code> method can be used to calculate the number of occurences of the specific element. </p>

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark", "Guanaco", "Vicuna", "Guanaco", "Camel")
print(my_tuple.count("Camel"))
```

<p><code>index()</code> method can be used to find the index of the specific element within the tuple. </p>

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark", "Guanaco", "Vicuna", "Guanaco", "Camel")
print(my_tuple.index("Camel"))
```

<code>len()</code> method can be applied to the tuple.

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark", "Guanaco", "Vicuna", "Guanaco", "Camel")
print(len(my_tuple))
```
