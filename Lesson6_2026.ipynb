<h1>Lesson 6: tuples</h1>

<ul><li>Tuples are one of 4 collection types available in python. (Others are dictionaries, lists and sets)</li>
    <li>Tuples are defined by the <code>()</code> brackets</li>
    <li>Tuples are unchangeable, meaning, once the tuple is created, it can not be changed.</li>
    <li>Elements within tuples have indexes.</li>
    <li>Tuples alllow duplicate values</li>
</ul>

<p>Acessing elements within the tuple and slicing the tuple is similar to that of list</p>

```python
my_tuple = ("Llama", "Camel", "Alpaca")
print(my_tuple[1])
```

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark")
print(my_tuple[1:3])
```

<p>Unlike with the list you can not just re-assign the value of the element, because tuples are unchangeable. The line below will not work.</p>

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark")
my_tuple[1] = "Zebra"
```

<p>But <b>You can convert the tuple to list</b> change it and then convert back</p>

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark")
my_list = list(my_tuple)
print(my_list)
my_list[1] = "Guanako"
convertedBack = tuple(my_list)
print(convertedBack)
```

<p>You can update the tuple using concatenation. This is similar to updating the string, technically what happens a new tuple is created and stored in the variable. </p>

```python
my_awsome_tuple = ("Llama", "Camel", "Alpaca")
my_awsome_tuple += ("Shark", )
print(my_awsome_tuple)                   
```

<p>Concatenating tuples</p>

```python
my_tuple = ("Llama", "Camel")
my_second_tuple = ("Alpaca", "Shark")
my_entire_tuple = my_tuple + my_second_tuple
print(my_entire_tuple)
```

<p><b>Unpacking tuples </b></p>
<p>The easiest ways to unpack the tuple is to create as many variables as there are elements in the tuple. See example below</p>

```python
my_awsome_tuple = ("Llama", "Camel", "Alpaca")
(first_elem, second_elem, third_elem) = my_awsome_tuple
print(first_elem)
```

<p>Unpacking tuple with asterisk <code>*</code>. Asterisk puts all the values that weren't allocated to the variables in a list.</p>

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark")
(first_elem, *other_elems, last_elem) = my_tuple
print(first_elem)
print(other_elems)
print(last_elem)
```

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark", "Guanaco", "Vicuna")
(first_elem, *other_elems, before_last_elem, last_elem) = my_tuple
print(first_elem)
print(other_elems)
print(before_last_elem)
print(last_elem)
```

<p>You can also loop through tuple.</p>

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark", "Guanaco", "Vicuna")
for elem in my_tuple:
    print(elem)
```

<h3>Tuple methods</h3>

<code>count()</code> method can be used to calculate the number of occurences of the specific element. 

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark", "Guanaco", "Vicuna", "Guanaco", "Camel")
print(my_tuple.count("Camel"))
```

<code>index()</code> method can be used to find the index of the specific element within the tuple. 

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark", "Guanaco", "Vicuna", "Guanaco", "Camel")
print(my_tuple.index("Camel"))
```

<code>len()</code> method can be applied to the tuple.

```python
my_tuple = ("Llama", "Camel", "Alpaca", "Shark", "Guanaco", "Vicuna", "Guanaco", "Camel")
print(len(my_tuple))
```
