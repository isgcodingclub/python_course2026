## Dictionaries ##
<blockquote> Dictionaries are used to store data values in <b> key:value </b> pairs. A dictionary is a collection which is ordered, changeable and does not allow duplicates. </blockquote><p>You can think about it, as about the freezer, the key is the tray number or name, and the value is what is stored in the tray.</p>

```python
# Below is the dictionary for conversion of the three letter code to one letter code
amino_acids = {"Gly": "G", "Ala": "A", "Val": "V"}
# Below is the dictionary that stores the amino acids and their molecular weights
amino_acids1 = {"G": 75.07, "A": 89.09, "V": 117.151}
```

To fetch out the item from the dictionary using its key you have to do the following:

```python
one_letter_code_ala = amino_acids1["A"]
print(one_letter_code_ala)
```

The dictionaries **can not have two items with the same key**. If you will try to add another item that has the same key as an item that is already ptresent in dictionary, the old item gets replaced by new one.

```python
thisdict ={"type":"VHH",
"Nr of cysteines": 2}
print(thisdict)
# let's add a duplicate
thisdict["Nr of cysteines"] = 3
print(thisdict)
```

<h4>Adding the key: value pairs to the existing dictionary. </h4>

<p>Getting the value from the dictionary using <code>dictionary.get()</code> method. This handles the situation when the key you are searching for is not present in the dictionary.</p>

The number of the key value pairs can be determined using the **len()** method.

<h3>Creating the dictionaty</h3>
<p>There are two ways how you can create the dictionary.
One is by using <code>dictionary = {}</code>. The second one is by using <code>dict()</code></p>

```python
dictionary1 = {"G": 75.07, "A": 89.09, "V": 117.151}
dictionary2 = dict(G = 75.07, A = 89.09, V = 117.151)
print(dictionary1)
print(dictionary2)
```

Constructing the dictionary out of the two lists, one of which will contain the keys and another will contain values. The methos in essence is similar to one you use whern you loop through the two lists simultaneously.

```python
list1 = ["Alpaca", "Llama", "Camel"]
list2 = ["Immunization1", "Immunization2", "Not useful"]
Result = dict(zip(list1, list2))
print(Result)
```

<h3>Concatentaing the dictionaries</h3>
<p>There are two ways how you can concatenate the dictionaries. One is <code>update()</code> that we covered before, another is unpacking <code>{**d1, **d2}</code> or using the <u>union operator</u> <code>|</code></code></p>

```python
d1 = {'G': 75.07, 'A': 89.09, 'V': 117.151}
d2 = {'C': 121.16, 'E': 147.13, 'Q': 146.15, 'G': 75.07}
d1.update(d2)
print(d1)
```

<p>Using union operator <code>|</code>. This workis in Python 3.9 +</p>

```python
d1 = {'G': 75.07, 'A': 89.09, 'V': 117.151}
d2 = {'C': 121.16, 'E': 147.13, 'Q': 146.15, 'G': 75.07}
merged_dict = d1 | d2
print(merged_dict)
```

<p>Using unpacking</p>

```python
d1 = {'G': 75.07, 'A': 89.09, 'V': 117.151}
d2 = {'C': 121.16, 'E': 147.13, 'Q': 146.15, 'G': 75.07}
merged_dict = {**d1, **d2}
print(merged_dict)
```

<h3>Dictionary methods</h3>

<p><code>dictionary.fromkeys()</code> method is a method for creation of the dictionary, when you want to quickly asign the same value to the different keys</p><p>Useful, when you create a dictionary that you will later update. For isntance you might be looping through the DNA sequences from the NGS data and calculating numbers of each nucleotide. Later you will use this information to estimate each nucleotide content in the sample. But in the beginning yiu need a dictionary to start with</p>

```python
my_keys = ["A", "T", "C", "G"]
value = 0
my_dict = dict.fromkeys(my_keys, value)
print(my_dict)
```

<code>dictionary.items()</code> method, that just lists the key value pairs. 

```python
amino_acids1 = {"G": 75.07, "A": 89.09, "Val": 117.151}
print(amino_acids1.items())
for key, value in amino_acids1.items():
    print(key, value)
```

<p>If you just want to fetch dictionary keys use <code>dictionary.keys()</code> method</p>

```python
amino_acids1 = {"G": 75.07, "A": 89.09, "Val": 117.151}
print(amino_acids1.keys())
```

<p>Similarly, if you want to know only values use <code>dictionary.values()</code></p>

```python
amino_acids1 = {"G": 75.07, "A": 89.09, "Val": 117.151}
print(amino_acids1.values())
```

<h4>Removing the key : value pair from the dictionary</h4>

<p>You can remove the item from the dictionary using the <code>dictionary.pop(keyname_to_be_removed)</code> method</p>

```python
my_dictionary = {'C': 121.16, 'E': 147.13, 'Q': 146.15, 'G': 75.07}
my_dictionary.pop('E')
print(my_dictionary)
```

<p>The <code>dictionary.pop()</code> actually returns the value from the removed key: value pair</p>

```python
my_dictionary = {'C': 121.16, 'E': 147.13, 'Q': 146.15, 'G': 75.07}
removed_value = my_dictionary.pop('E')
print(my_dictionary)
print(removed_value)
```

<p>You can also pass the optional argument into the method, <code>dictionary.pop(key_to_remove, defaultvalue)</code> where the <code>defaultvalue</code> is the value that returns if the element with the specified key does not exist.</p>

```python
my_dictionary = {'C': 121.16, 'E': 147.13, 'Q': 146.15, 'G': 75.07}
removed_value = my_dictionary.pop('T', "Does not exist")
print(removed_value)
print(my_dictionary)
```

<p>Removing the pecific key:value pair using <code>del</code> method</p>

```python
my_dictionary = {'C': 121.16, 'E': 147.13, 'Q': 146.15, 'G': 75.07}
del my_dictionary["C"]
print(my_dictionary)
```

<p>You can delete the entire contrents of the dictionary using<code>dictionary.clear()</code></p>

```python
my_dictionary = {'C': 121.16, 'E': 147.13, 'Q': 146.15, 'G': 75.07}
my_dictionary.clear()
print(my_dictionary)
```

<h3>Other useful methods</h3>

<p>The<code>setdefault()</code> method can be used to update dictionary</p><p>The synthax is <code>dictionary.setdefault(keyname, value)</code>. If the key exists in the dictionary, then the value is fetched. If the key does not exist in the dictionary, the user-provided value is fetched and the dictionary gets updated. </p>

```python
# If the key exists in the dictionary
my_dictionary = {'C': 121.16, 'E': 147.13, 'Q': 146.15, 'G': 75.07}
result = my_dictionary.setdefault('C', 100)
print(my_dictionary)
print(result)
```

```python
# if it does not exits
my_dictionary = {'C': 121.16, 'E': 147.13, 'Q': 146.15, 'G': 75.07}
result = my_dictionary.setdefault('X', 100)
print(result)
print(my_dictionary)
```

<p><code>dictionary.copy()</code>method just copies the dictionary</p>

```python
my_dictionary = {'C': 121.16, 'E': 147.13, 'Q': 146.15, 'G': 75.07}
my_cloned_dictionary = my_dictionary.copy()
print(my_cloned_dictionary)
```

<h3>Note on the dictionary view objects 🪟</h3>

<p>The <code>keys()</code> and <code>.values()</code> are the view obejcts, meaning that they are dynamic. Which I am going to demonatrate below: </p>

```python
my_dictionary = {'C': 121.16, 'E': 147.13, 'Q': 146.15, 'G': 75.07}
# save the keys in a variable
my_keys = my_dictionary.keys()
print(my_keys)

# update the dictionary using del, remove one of the key : value pairs
del my_dictionary["E"]
# print keys second time
print(my_keys)
```

<p>In the example above we created the variable called <code>my_keys</code> that stored the keys at the given moment. We printed it. Then we updated the dictionary, and the variable got updated in the background. That's why it is called view object. You can think about it as about the window through which you can see what happens with the keys. </p>
