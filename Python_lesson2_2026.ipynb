👉 **The weird behaviour of the float numbers**<p>
<p>The example below is </p>

```python
x = 0.29*100
print(x)
```

If you multiply 0.29 by 100 you get 28.999999999996. To explain this issue we need to understand how the computers actually work. </p> <p> Humans are used to the decimal system. Lets consider the decimal number 2024. It is equal to $$2024 = 2*1000 + 0*100 + 2*10 + 4*1$$ $$2024 = 2*10^3 + 0*10^2 + 2*10^1 + 4*10^0$$ </p> <p> In the binary system the numbers are defined similarly, but instead of having 1-9 options for each space you have only two 0 or 1 and each position in a power of 2. Let's consider example below:
$$ 1011$$ - this is number 11 in binary system.
$$1011 = 1*2^3 + 0*2^2+ 1*2^1 + 1*2^0 = 8+0+2+1 = 11 $$
Why I am showing you that? Because the fractions are also represented as the sums of the 2s raised to power of something. Lets consider 0.25:
$$ 0.25 = \frac{4}{10} =  \frac{1}{4} =  \frac{1}{2^2} = 2^{-2} $$
Let's convert 0.25 into the binary system. 
$$ 0.25 = \frac{1}{2^2} = 0*2^0 + 0*2^{-1} + 1*2^{-2}$$
So binary representation would be $$0.01$$
<p>But not all fractions can be repesented in this way.</p> 
<h3>Proof that the 0.29 can not be converted into the binary format</h3>
Let's try to convert the 0.29 into the bindary format. To be able to do that, it is necessary to represent the number as a sum of the 2s raised to power. 
<p>If you think about any fraction number that is a sum of the powers of 2 it can be represented as $$ \frac{N}{2^k} $$</p>
So let's assume that $$ 0.29 = \frac{N}{2^k} $$ where N and k are integer numbers. <p>Then $$\frac{29}{100} = \frac{N}{2^k}$$</p>
<p>That would imply that $$N*100 = 29*2^k$$</p>
<p>Which is not possible because you will newer get the number ending with 100 by multiplying a 29 by intereger, that is based on the power of two. This is because all numbers obtained by raising 2 to power $$2^k$$ are ending with 2, 4, 6 or 8 and it is not possible to multiply any of these by 9 and get 0.</p>
<p style = "background-color:#e9d5f2;">So there are fractions that can not be repersented in binary format, so computers store not the exact value, but the nearest value that can be stored in the binary format.</p><p>And thereore when we are multiplying 0.29*100 does not actually multiply 0.29, but the closest number that is possible to store as binary number</p>
