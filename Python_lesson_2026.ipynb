### Strings and variables ###

The string is defined by the `""` or `''` quotation marks. Either are fine, so you can do either `"this is my string"` or `'this is my string'` but you can not mix two types of quotation marks. Let's print some strings:

```python
print("The VHH company")
print('The VHH company')
```

As I already wrote above, you <u> can not mix different quotation marks</u>. The example below will not work.

```python
print("The VHH company')
```

### String concatenation

The string can be concatenated <u> to another string </u>, but not to number:


```python
print("this"+ " string")
```

```python
print("This string" + 42)
```

The way how to circumvent this is either putting number in the brackets "42", so that it becomes a string or use str()

```python
print("This string " + str(42))
print("This string " + "42")
```

You <b> can not substract </b> one string from another

```python
print("string" - "ring")
```

You can not multiply or divide one string by another

```python
print("string"*"ring")
```

```python
print("string"/"ring")
```

You can multiply string by number, it will just repeat the string

```python
print("string"*3)
```

<b> Slicing the string </b> <br>
<b>Using the <code>string[start:stop]</code> method</b>

download (1).png

If you want to slice out the single digit:

```python
s = "Isogenica"
print(s[2])
print(s[-2])
```

The synthaxis is <b><code>string[start:end]</code></b> but the the digit of index end is not included. Hence in the example below the slice <code>Isogenica[0:3]</code> will generate <code>Iso</code> string. 

```python
s = "Isogenica"
print(s[0:3])
```

Thus, it might be more correct to say that the synthax for slicing part of the string is <b><code>string[start:end+1]</code></b> and if you want to include the digit witj the index 3 you need to include 4 as your slicing end. 

```python
s = "Isogenica"
print(s[0:4])
```

Similraly you can use the <b> reverse numbering </b>.

```python
s = "Isogenica"
print(s[-7:-4])
```

However, the slicing will only work if you will slice in the <b>left to right direction</b>. The example below generates the empty string. 

```python
s = "Isogenica"
print(s[-7:-9])
print(len(s[-7:-9]))
```

You can mix two types of the numbering as long as you are going in <b>left to right direction.</b>

```python
s = "Isogenica"
print(s[-8:4])
```

<b>Using semicolon for slicing</b>
<p>Using the semicolon for slicing, useful when you do not know exact length of the string.</p>
<p>The semicolon means either "from the beginning" or "till the end".</p> 
<p>The digit with index <b>end</b> is not included</p>

```python
s = "Isogenica"
print(s[3:]) 
print(s[:2])
```

<b> Using [start:stop:step] method. </b> <br> To illustrate what about this method I will use 01234567 string. <br> start = 0, stop = 5. So it selects 01234 string and then selectes the diigit that is separated by "step" from the start of the selection. So, 0, 2, 4. The end digit is not included. 


```python
s = "01234567"
print(s[0:5:2])
```

<p>Another example that is designed to show that the end index is not included. In the example below it is possible to reach the digit "6" if your starting position is 0 and step is 2, but it is not printed because "6" corresponds to the end position</p>

```python
s = "01234567"
print(s[0:6:2])
```

<b>Start stop step method</b> actually works backwards.

```python
string = "0123456789"
print(string[-1:-10:-2])
```

<p>Cool thing you can do using slicing: invert the string.</p>
<p>You remeber that the semicolon <code>:</code> means from beginning or till the end.</p>
<p>So the construction <code>string[::]</code> means from beginnning till the end, but it is not very useful</p>

```python
string = "Isogenica"
print(string[::])
```

<p>The example shown below is technically a string slicing from beginning till the end in with the step -1.</p>
<p>It results in the inversted string</p>

```python
string = "Isogenica"
string_new = string[::-1]
print(string_new)
```

<h4><code>.find()</code> and <code>.index()</code> methods</h4>

These methods can be used for finding the index of the specific element withing the string. Both **.find()** and **.index()** find the first instance of the element you want to find.

```python
string = "Butterfly"
print(string.find("fly"))
```

<p>The <code>.find()</code> method is only capable of finding the first instance of the searched string</p>

```python
string = "Butterfly, I can fly"
print(string.find("fly"))
```

<p><b>Below is the example more relevant to us</b></p>
<p>In the example below the indexes of the framework 1 and framework 2</p>

```python
sequence = "EVQLVESGGGLVQAGDSLRLSCTASGRTFSRAVMGWFRQAPGKEREFVAAISAAPGTAYYAFYADSVRGRFSISADSAKNTVYLQMNSLKPEDTAVYYCAADLKMQIAAYANQRSVDYWGQGTQVTVSS"
framework_1 = "EVQLVESGGGLVQAGDSLRLSC"
framework_2 = "WFRQAPGKEREFV"
framework1_start = sequence.find(framework_1)
framework2_start = sequence.find(framework_2)
print(framework1_start)
print(framework2_start)
```

If the **.find()** method is used to find something that does not exist in the sting, it returns -1.

```python
string = "The string"
print(string.find("Alpaca"))
```

**.index()** method works similarly to the **.find()** method, but when it is ordered to use the substring that does not exist in the string it gives an error.

```python
string = "The string"
print(string.index("ring"))
```

```python
print(string.index("Alpaca"))
```
