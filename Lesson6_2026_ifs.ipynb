**Input Statement**

Input Statement - **The input()** function allows user input into the code. The resulting input can be stored in a **variable**.

````python
print('Enter the codon for methionine')
codon = input()
print('The codon you entered was ' + '-'+ codon + '-')
````

Input statements can also a prompt text if text is added within the **()**

````python
print('Enter the codon for methionine')
codon = input('Enter the codon here: ')
print('The codon you entered was ' + '-'+ codon + '-')
````

**input()** statement entries (and subsequently its variable) are **ALWAYS** treated as strings by default !

````python
print('Enter the codon for methionine')
codon = input()
print(type(codon))
````

<p>The code below is going to generate the error, because the input is the string and we are trying to concatenate it with a number</p>

````python
print('Enter a number of sequences')
codon = input()
print(codon + 34)
````

<p>So instead of doing that, we should convert string to insteger, using <code>int()</code>. 

````python
print('Enter a number of sequences')
codon = input()
print(int(codon) + 34)
````

### If statements ###
The simplest <b> if statement </b> will look like this pseudocode:
```
if condition:
    do something
```
<p>The important bit is the intendation, after if statement and the semicolon the next line starts with the <b>tab</b>.</p>
<p>The result of the condition is boolean. It is <code>True</code> if condition is satisfied. And <code>False</code> if it is not satisfied.</p>

````python
a = 6
if a%2 == 0:
    print("Even number")
````

<p> Below is the expression that shows that the condition from the if statement above results in Boolean value</p>

````python
a = 6
print(a%2 == 0)
````

<p>So in other words the if statement says </p>

```
if condition == True:
    do something
```

The more complicated <b> if statement </b> would include alternative condition and using <b> elif </b>:
```
if condition:
    do something
elif another_condition:
    do_something_else
```

````python
CDR1 = "AASARTFYTMG"
if len(CDR1) == 11:
    print("correct CDR1 length")
elif len(CDR1) != 11:
    print("incorrect CDR1 length")
# != means not equal == means equal when searching for specific thing.
````

Finally, you can add the <b> else </b> statement to capture all the situations when all of your conditions are not valid

```
if condition1:
    do something
elif condition2:
    do_something1
elif condition3:
    do_something2
else:
    do something else
```

<b> pass, continue, break and exit statements. </b> <br> How to do nothing in pyton.


<p><b>1. Pass </b> statement or how to do nothing in python </p>

````python
CDR1 = "AASARTCYTMG"
for position, letter in enumerate(CDR1):
  if letter!= "C":
    pass
  elif letter == "C":
    print("There is a cysteine at "+ str(position+1))
````

<b> Continue </b> statement switches to the next iteration.

````python
list_of_lentgths = [1, 3, 5, 0, 13, 14]
for i in list_of_lentgths:
  if i == 0:
    continue
  print(i)
````

<b> continue </b> forces the loop to start at the next iteration, whereas <b> pass</b> means, “there is no code to execute here,” and it will continue through the remainder of the loop body. <br> In the example below it still printed the zero </br>

````python
list_of_lentgths = [1, 3, 5, 0, 13, 14]
for i in list_of_lentgths:
  if i == 0:
    pass
  print(i)
````

<b> Break </b> statement breaks avay from the loop and winishes the loop when it encounters specific condition.

````python
CDR1 = "AASARTCYTMG"
for position, letter in enumerate(CDR1):
  print(letter)
  if letter== "C":
    break
print("Loop went up to C")
````

<p>If you have a nested loop the break statement is goung to break out of the innermost loop</p>

````python
for i in range(1, 10):
    for y in range(1, 10):
        if y == 5:
            print("Reached sub-chapter 5. Will continue to the next outer loop inetration")
            break
        print("Chapter" + str(i) + "." + str(y))
        
````

**Exit** statement not only breaks from the loop but also finishes all the code and closes the entire script. <b> This is not what is happening in this compiler. </b>

````python
CDR1 = "AASARTCYTMG"
for position, letter in enumerate(CDR1):
  print(letter)
  if letter == "C":
    exit()
print("Loop went up to C")
````

Range is used to create the sequence of numbers that can be used in the loop.

<b>If statemets and list comprehension</b>

<p>First, let's revise how the simple if statement in the list comprehension will look like.</p>
<p>Below is the list of numbers, we set up the conditions so that only even numbers remain in the list.</p>

````python
list_numbers = list(range(0, 11))
print("below is our list of numbers")
print(list_numbers)

even_numbers = [a for a in list_numbers if a%2 == 0]
print("below is the list of selected even numbers")
print(even_numbers)
````

<p>You can also have the list comprehension with <code>else</code>. The synthaxis is slightly more complicated</p>

```
result = [do something if condition else do something else for a in collection]
```

<p>Let's construct the example in which we multiplly the even numbers by two and the odd numbers by three.</p>

````python
list_numbers = list(range(0, 11))
resultant_list = [a*3 if a%2 else a*2 for a in list_numbers]
print(resultant_list)
````
